```python
import pandas as pd

# Specify the file path
excel_file_path = '/media/DATA/martina_ma/pathologic_fractures_clean.xlsx'

# Read the Excel file into a pandas DataFrame
# Read a specific sheet by name
df = pd.read_excel(excel_file_path, sheet_name='data')


# Display the DataFrame
print(df)
```

```python
from BIDS.core.vert_constants import v_name2idx, v_idx2name

print(v_idx2name)
print(v_name2idx)

class Anomaly:
    def __init__(self, df_row):
        self.name = df_row['ID'].lower()
        self.sex = df_row['sex']
        self.label = df_row['label']
        self.age = df_row['age']
        self.fracture_heights = df_row['fracture_heights'].replace(" ", "").split('_')
        self.fracture_heights_mapped= [v_name2idx.get(value, -1) for value in  df_row['fracture_heights'.replace(" ", "")].split('_')]#.map(v_name2idx)
        self.main_fracture = df_row['main_fracture']
        self.exam_date = df_row['exam_date']

anomaly_dict = {}
# Iterate over each row in the DataFrame and create a Person instance
for index, row in df.iterrows():
    person_id = row['ID']
    if person_id not in anomaly_dict:
        anomaly_dict[person_id] = Anomaly(row)

# Now, you can access the attributes for each person
for person_id, person_instance in anomaly_dict.items():
    print(f"ID: {person_instance.name}, Sex: {person_instance.sex}, Label: {person_instance.label}, Age: {person_instance.age},'F_H:{person_instance.fracture_heights},'F_H_m:{person_instance.fracture_heights_mapped}")
```
